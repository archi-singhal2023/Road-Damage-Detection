```python
import os

root = "dataset"      

print("Train Images:", len(os.listdir(f"{root}/train/images")))
print("Train Labels:", len(os.listdir(f"{root}/train/labels")))
print("Valid Images:", len(os.listdir(f"{root}/valid/images")))
print("Test Images:", len(os.listdir(f"{root}/test/images")))
```

```python
from ultralytics import YOLO
```

```python
# Load YOLOv8 small model
model = YOLO("yolov8s.pt")
```

```python
import torch
print(torch.cuda.is_available())      # Should be True
print(torch.cuda.get_device_name(0))  # Your GPU
```

```python
# Train the model
model.train(
    data="rdd.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    name="rdd_yolov8",
    device = "cuda" if torch.cuda.is_available() else "cpu",   
    workers=1,
    patience = 3,
    amp = False
)
```

```python
# Load last checkpoint
model = YOLO("runs/detect/rdd_yolov8/weights/last.pt")

# Resume training
model.train(
    resume=True
)
```

```python
# Load the best trained model
model = YOLO("runs/detect/rdd_yolov8/weights/best.pt")
```

```python
# Evaluate the model on the validation set
metrics = model.val()
```

```python
metrics
```

```python
# Evaluate model on TEST set
test_metrics = model.val(
    data="rdd.yaml",
    split="test",
    plots=True
)
```
